### In this project, I take raw data from Uber in the form of a CSV file and convert it to a dataframe. I utilize the Python libraries, Pandas and Seaborn, to analyze rides in New York City during the month of April 2014.  

```python
%pylab inline 
import pandas 
import seaborn 
```

# Load CSV file into memory 

```python
data = pandas.read_csv("Desktop/uber-raw-data-apr14.csv")
```

```python
data
```

```python
data["Date/Time"] = data["Date/Time"].map(pandas.to_datetime) h
```

## Convert datetime and add useful columns 

```python
data["Date/Time"] = data["Date/Time"].map(pandas.to_datetime)
```

```python
data.head()
```

```python
data.tail()
```

## Day of month

```python
def get_dom(dt):
    return dt.day
data['dom'] = data['Date/Time'].map(get_dom)
```

```python
def get_weekday(dt):
    return dt.weekday()
data["weekday"] = data["Date/Time"].map(get_weekday)

def get_hour(dt):
    return dt.hour
data["hour"] = data["Date/Time"].map(get_hour)

data.tail()
```

# Analysis

## Analyze DoM 

```python
hist(data.dom, bins=30, rwidth = .8, range = (0.5, 30.5))
xlabel("Data of Month")
ylabel("Frequency")
title("Frequency by DoM - Uber - Apr 2014")
```

```python
#for k, rows in data.groupby('dom'):
    #print(k, rows)
    #break
    #print(k, len(rows))
def count_rows(rows):
    return len(rows)
by_date = data.groupby('dom').apply(count_rows)
by_date
```

```python
bar(range(1,31), by_date)
```

## Sort frequency of dates  

```python
by_date_sorted = by_date.sort_values()
by_date_sorted
```

```python
bar(range(1,31), by_date_sorted)
xticks(range(1,31), by_date_sorted.index)
;
```
